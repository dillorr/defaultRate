## predict proprietary

```python
aucProprietaryLabels = ('log', 'knn', 'n-bayes', 'svm', 'd-tree', 'r-forest')
aucProprietaryValues = (0.60, 0.81, 0.92, 0.59, 0.99, 0.98)
```

```python
aucProprietary = pd.DataFrame(list(zip(aucProprietaryLabels, aucProprietaryValues)), columns = ['algorithm', 'auc'])
```

```python
import seaborn as sns

aucProprietary = aucProprietary.sort_values(by = 'auc', ascending = False).reset_index()

sns.set(style = 'darkgrid', font = 'serif')
g = sns.barplot(x = 'algorithm', y = 'auc', data = aucProprietary)
plt.title('Performance of Algorithms Predicting School Type')

#label values
for index, row in aucProprietary.iterrows():
    g.text(row.name, row.auc, round(row.auc, 6), color='black', ha="center")
```

## predict high/low cdr

```python
aucCDRLabels = ('log', 'knn', 'n-bayes', 'svm', 'd-tree', 'r-forest')
aucCDRValues = (0.64, 0.59, 0.77, 0.60, 1.00, 1.00)
```

```python
aucCDR = pd.DataFrame(list(zip(aucCDRLabels, aucCDRValues)), columns = ['algorithm', 'auc'])
```
