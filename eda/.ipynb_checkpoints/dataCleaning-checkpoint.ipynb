```python
# import df without missing vals
df = pd.read_csv(r'/Users/dillorr/defaultRate/dillon/defaultRate/data/withoutNulls.csv', encoding = "ISO-8859-1")
df.head()
```

```python
# make copy of dataset for manipulation
dRateCopy = df.copy()

# replace numbered values for schoolType with strings
dRateCopy.loc[(dRateCopy.schooltype == 1),'schooltype']='public'
dRateCopy.loc[(dRateCopy.schooltype == 2),'schooltype']='private'
dRateCopy.loc[(dRateCopy.schooltype == 3),'schooltype']='proprietary'
dRateCopy.loc[(dRateCopy.schooltype == 5),'schooltype']='foreign'
dRateCopy.loc[(dRateCopy.schooltype == 6),'schooltype']='foreign'
dRateCopy.loc[(dRateCopy.schooltype == 7),'schooltype']='foreign'

# remove foreign schools
dRateCopy = dRateCopy[dRateCopy.schooltype != 'foreign']

dRateCopy.schooltype.head()
```

```python
# get df with 1 hot encoding for schoolType
withDummies = pd.concat([dRateCopy, dRateCopy.schooltype.str.get_dummies()], axis=1)

withDummies
```

```python
# make copy of dataset for manipulation
withDummiesCopy = withDummies.copy()

# replace numbered values for program lenth with strings

# 0—Short-Term (300–599 hours)
withDummiesCopy.loc[(withDummies.proglength == 0),'proglength']='shortterm'

# 1—Graduate/Professional (≥ 300 hours)
withDummiesCopy.loc[(withDummies.proglength == 1),'proglength']='gradprofessional'

# 2—Non-Degree (600–899 hours)
withDummiesCopy.loc[(withDummies.proglength == 2),'proglength']='nondegree'

# 3—Non-Degree 1 Year (900–1799 hours)
withDummiesCopy.loc[(withDummies.proglength == 3),'proglength']='nondegree1year'

# 4—Non-Degree 2 Years (1800–2699 hours)
withDummiesCopy.loc[(withDummies.proglength == 4),'proglength']='nondegree2years'

# 5—Associate's Degree
withDummiesCopy.loc[(withDummies.proglength == 5),'proglength']='associates'

# 6—Bachelor's Degree
withDummiesCopy.loc[(withDummies.proglength == 6),'proglength']='bachelors'

# 7—First Professional Degree
withDummiesCopy.loc[(withDummies.proglength == 7),'proglength']='firstprofessional'

# 8—Master's Degree or Doctor's Degree
withDummiesCopy.loc[(withDummies.proglength == 8),'proglength']='mastersordoctors'

# 9—Professional Certification
withDummiesCopy.loc[(withDummies.proglength == 9),'proglength']='professionalcertification'

# 10—Undergraduate (Previous Degree Required)
withDummiesCopy.loc[(withDummies.proglength == 10),'proglength']='undergradprevdegreereq'

# 11—Non-Degree 3 Plus Years (≥ 2700 hours)
withDummiesCopy.loc[(withDummies.proglength == 11),'proglength']='nondegree3years'

# 12—Two-Year Transfer 
withDummiesCopy.loc[(withDummies.proglength == 12),'proglength']='2yeartransfer'


withDummiesCopy.proglength
```

```python
# get df with 1 hot encoding for schoolType
withDummies = pd.concat([withDummies, withDummiesCopy.proglength.str.get_dummies()], axis=1)

withDummies['proglength'] = withDummiesCopy['proglength']
withDummies['proglength']
```

```python
variance = withDummies.var()
variance[variance.values == 0]
```

```python
# remove zero variance variables
dfReduced = withDummies.drop(columns = ['year1', 'year2', 'year3', 'military'], axis = 1)
dfReduced.columns
```

```python
# make copy of dataset for adding ethnic code dummy variables
dfReducedCopy = dfReduced.copy()

# replace numbered values for ethnic code for strings

# 1—Native American
dfReducedCopy.loc[(dfReducedCopy.ethniccode == 1),'ethniccode']='nativeAmericanCollege'

# 2—HBCU
dfReducedCopy.loc[(dfReducedCopy.ethniccode == 2),'ethniccode']='hbcuCollege'

# 3—Hispanic
dfReducedCopy.loc[(dfReducedCopy.ethniccode == 3),'ethniccode']='hispanicCollege'

# 4—Traditionally Black College
dfReducedCopy.loc[(dfReducedCopy.ethniccode == 4),'ethniccode']='blackCollege'

# 5—Ethnicity Not Reported 
dfReducedCopy.loc[(dfReducedCopy.ethniccode == 5),'ethniccode']='notReportedCollege'

dfReducedCopy.ethniccode
```
