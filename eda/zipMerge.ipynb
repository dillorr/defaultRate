```python
# small CDR File
cdrSmall = pd.read_csv(r'/Users/dillorr/defaultRate/dillon/defaultRate/data/merged/excelTrimmed/cdrSmall.csv', encoding = "ISO-8859-1")
cdrSmall.describe()
```

```python
# large CDR file
cdrLarge = pd.read_csv(r'/Users/dillorr/defaultRate/dillon/defaultRate/data/merged/excelTrimmed/cdrLarge.csv', encoding = "ISO-8859-1")
cdrLarge.describe()
```

```python
combinedCDR = pd.concat([cdrLarge, cdrSmall], sort = False)
```

```python
combinedCDR.describe()
```

```python
combinedCDR.columns
```

```python
# remove newlines and spaces
combinedCDR.columns = combinedCDR.columns.str.replace("\n", "")
combinedCDR.columns = combinedCDR.columns.str.replace(" ", "")
combinedCDR.columns = combinedCDR.columns.str.replace("_x", "")
combinedCDR.columns = combinedCDR.columns.str.replace("_y", "")
combinedCDR.columns = combinedCDR.columns.str.replace("_", "")
combinedCDR.columns = combinedCDR.columns.str.lower()
combinedCDR = combinedCDR.sort_values(by='opeid')
combinedCDR.head()
```

```python
# export csv
combinedCDR.to_csv('/Users/dillorr/defaultRate/dillon/defaultRate/data/merged/CDRCombined.csv', index=False)
```
